**Thrift schema (student.thrift)**

```python
%%writefile ../schema/student.thrift

struct Student {
  1: required string name,
  2: required i32 age,
  3: required list<string> courses
}

service School {
  Student enrollCourse(1: required Student student, 2: required string course)
}
```

**Thrift server (student_thrift_server.py)**

```python
%%writefile ../student_thrift_server.py
import thriftpy2
from thriftpy2.rpc import make_server

# Load the Thrift schema
student_thrift = thriftpy2.load("./schema/student.thrift", module_name="student_thrift")

class SchoolHandler:
    # This method name must match what you wrote in the service definition
    def enrollCourse(self, student, course):
        # Add the course into the list, then return the updated record
        student.courses.append(course)
        return student

server = make_server(student_thrift.School, SchoolHandler(), client_timeout=None)

print("Starting and running the server...")
print("Press Ctrl+C to stop the server.")
server.serve()
```

**Thrift client (student_client.py)**

```python
import thriftpy2
from thriftpy2.rpc import make_client

student_thrift = thriftpy2.load("../schema/student.thrift", module_name="student_thrift")

# Create a client for the School service
school = make_client(student_thrift.School, timeout=None)

# Create a Student record
alice = student_thrift.Student(name="Alice", age=20, courses=["math"])

# Call RPC: enrollCourse(student, course)
alice = school.enrollCourse(alice, "history")

# Check updated courses
alice.courses
```
